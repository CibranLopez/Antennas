```python
coefficients = [1, 0.201, 1.171, 1.710, 1.835, 2.029, 2.413, 2.625, 2.413, 2.029, 1.835, 1.710 , 1.171, 0.201, 1]

a = np.roots(coefficients)

for v in a:
    print(cmath.polar(v))
```

```python
def dolph_chebyshev(w, c):
    return c[0] + c[1]*w + c[2]*w**2 + c[3]*w**3 + c[4]*w**4 + c[5]*w**5 + c[6]*w**6 + c[7]*w**7 + c[8]*w**8 + c[9]*w**9 + c[10]*w**10 + c[11]*w**11 + c[12]*w**12 + c[13]*w**13 + c[14]*w**14

c = np.arange(-1, 1+0.001, 0.001)
f = dolph_chebyshev(c, coefficients)
f = 20 * np.log10(f)
plt.plot(c, f)
```
